# How to compare two data's  no_code and real label

```python
df['gender'].value_counts()
```

```python
genderlist=df['gender'].value_counts(sort=False,normalize=False,bins=2)
genderlist
```

-----------------------

```python
genderlist=list(genderlist)
genderlabel=["Male","Female"]
plt.style.use("fivethirtyeight")
plt.pie(genderlist,labels=genderlabel,wedgeprops={'edgecolor': 'white'},shadow=True,autopct='%1.1f%%')
plt.title("Customer Gender Distribution")
```

```python
agelist=df['age'].value_counts(sort=False,normalize=False,bins=4)
```

```python
#  radius
plt.style.use("fivethirtyeight")
agelist=list(agelist)
agelabel=['<20',"20-29",'30-39',"40 and above"]
plt.pie(agelist,labels=agelabel,autopct='%1.1f%%',radius=1,pctdistance=1,labeldistance=1.3,wedgeprops=dict(linewidth=2,width=0.3,edgecolor='w'))
plt.title("Customer Age Distribution")
```

```python
df['status'].value_counts()
```

```python
statuslist=df['status'].value_counts(sort=False,normalize=False,bins=4)
statuslist
```

```python
statuslist=(statuslist)
statuslabel=["Student","Self-employed","Employed","Housewife"]
plt.pie(statuslist,labels=statuslabel,wedgeprops={'edgecolor': 'white'},shadow=True,autopct='%1.1f%%')
plt.title("Customer Status Distribution",fontsize=20)
```

# 
## df_num.iloc[:,7]

```python
df_num.iloc[:,6].value_counts()
```

```python
df_num.iloc[:,6].replace(["never","I dont like coffee","Never","Never "],"Never buy",inplace=True)
df_num.iloc[:,6]=df_num.iloc[:,6].fillna(value="Never buy")
methodlist=df_num.iloc[:,6].value_counts()
methodlist
```

```python
methodlist=list(methodlist)
methodlabel=list(df_num.iloc[:,6].unique())
plt.pie(methodlist,labels=methodlabel,wedgeprops={'edgecolor': 'white'},shadow=True,autopct='%1.1f%%')

plt.title("Customer Purchasing Method Distribution",fontsize=20)
```

```python
timespend=pd.crosstab(df_num.iloc[:,6],df_num.iloc[:,7])
timespend
```

```python
timespend.plot(kind="line",figsize=(12,5))
plt.xticks(rotation=0)
plt.legend(bbox_to_anchor =(1, 1.15), ncol = 1) 

plt.title("Customer Purchasing Time Spend Compare",fontsize=30)
```

```python
timespend.plot(kind="bar",figsize=(12,5))
plt.xticks(rotation=0)
plt.legend(bbox_to_anchor =(1, 1.15), ncol = 1) 

plt.title("Customer Purchasing Time Spend Compare",fontsize=30)
```

```python
yearchoose=pd.crosstab(df['age'], df['chooseRate'], normalize='index')
yearchoose=yearchoose.rename(index={0:"Below 20",1:"20-29",2:"30-39",3:"40 Above"})
yearchoose
```

```python
yearchoose.plot(kind="bar",figsize=(10,4))
plt.xticks(rotation=0)
plt.legend(bbox_to_anchor =(1, 0.75), ncol = 1) 
plt.title("Customer Age and Choose Rate Compare",fontsize=25)

yearchoose.plot(kind="bar",stacked=True)
plt.legend(bbox_to_anchor =(1, 1.15), ncol = 1) 
plt.xticks(rotation=0)

plt.title("Customer Age and Choose Rate Compare",fontsize=20)
```
