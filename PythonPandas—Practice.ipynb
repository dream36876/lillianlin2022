```python
ins=pd.read_csv("insurance.csv")
ins.head()
```

```python
ins["region"].value_counts()
```

# Aggregating method

```python
ins.groupby("sex").agg({"charges":"mean"}).round(-3)
```

```python
ins.groupby("sex").agg({"charges":["mean","count","var"]}).round(0)
```

# Mutiple Grouping columns

```python
ins.groupby(["sex","region"]).agg({"charges":["mean","count","max"]}).round(0)
```

```python
pt=ins.pivot_table(index=["sex","region"],
                  values="charges",aggfunc=["mean","count","max"]).round(0)
pt
```

```python
pt=ins.pivot_table(index="sex",columns="region",
                  values="charges",aggfunc="mean").round(0)
pt
```

# Styling pivot tables

```python
pt.style.highlight_max()
```

```python
pt.plot(kind="bar",figsize=(10,4))
```
